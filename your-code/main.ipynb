```python
x = np.arange(0,100)
y = x*2
z = x**2
```

#### Plot (x, y) and (x, z).
There are 2 ways of doing this. Do in both ways.

#### Use your previous code but now, resize your plot.
**Hint**: Add the figsize argument in plt.subplots().

If you want, try to add a title to the plot or even axes labels. You can also play with the fontweight and fontsize of the titles and labels. 

```python
# resixe with figsize
# add title with set_title
# change size with fontsize
# change style with fontstyle

fig, ax = plt.subplots(1,2, figsize=(16,5))
ax[0].plot(x,y)
ax[1].plot(x,z)

ax[0].set_title('X vs Y', fontsize=17, fontstyle='italic')
ax[1].set_title('X vs Z',fontsize=17, fontstyle='italic')

ax[0].set_xlabel('X')
ax[0].set_ylabel('Y')

ax[1].set_xlabel('X')
ax[1].set_ylabel('Z')

fig.suptitle('X comparision', fontsize=25, fontname='Sans', fontweight='bold')
```

#### Plot both $y=x^2$ and $y=e^x$ in the same plot using normal and logarithmic scale.
**Hint**: Use `set_xscale` and `set_yscale`.

```python
# instead of using y variables, introduce graphs by using x and f(x)
# change scale with set_yscale

fig, ax = plt.subplots(1,2, figsize=(20,5))
ax[0].plot(x, x**2, x, np.exp(x))
ax[0].set_yscale('linear')
ax[0].set_xlabel('x')
ax[0].set_ylabel('y')
ax[0].set_title('Normal scale', fontweight='bold', fontsize=21)

ax[1].plot(x, x**2, x, np.exp(x))
ax[1].set_yscale('log')
ax[1].set_xlabel('x')
ax[1].set_ylabel('y')
ax[1].set_title('Logaritmic scale', fontweight='bold', fontsize=21)
```

#### As a bonus challenge, try to add a legend to the plot.

```python
# Introduce legend wit legend()

fig, ax = plt.subplots(1,2, figsize=(20,5))
ax[0].plot(x, x**2, x, np.exp(x))
ax[0].set_yscale('linear')
ax[0].set_xlabel('x')
ax[0].set_ylabel('x**2')
ax[0].set_title('Normal scale', fontweight='bold', fontsize=21)
ax[0].legend(['x**2','e**x'])

ax[1].plot(x, x**2, x, np.exp(x))
ax[1].set_yscale('log')
ax[1].set_title('Logaritmic scale', fontweight='bold', fontsize=21)
ax[1].legend(['x**2','e**x'])
```

#### From the Fitbit data, we want to visually understand:

How the average number of steps change by month. Use the appropriate visualization to show the median steps by month. Is Fitbitter more active on weekend or workdays? All plots must be in the same jupyter notebook cell.

**Hints**:

* Use Months_encoded and Week_or Weekend columns.
* Use matplolib.pyplot object oriented API.
* Set your size figure to 12,4
* Explore plt.sca
* Explore plt.xticks
* Save your figures in a folder called `figures` in your repo. 

```python
# Create two subplots

fig, ax = plt.subplots(1,2, figsize = (12,4))

# MEDIAN STEPS BY MONTH_ENCODED
# Group by month_encoded and aggregate the average of the steps
# Create a bar graph with its title, labels and horizontal line idnicating the overall average
steps_month = round(fitbit.groupby('Months_encoded')['Steps'].mean(), 2)
steps_month.plot(kind='bar', ax=ax[0], color=['royalblue', 'forestgreen'])
ax[0].set_title('Average Steps Per Month', fontsize=15, fontweight='bold')
ax[0].set_xlabel('Month')
ax[0].set_ylabel('Avg Steps')
ax[0].axhline(steps_month.mean(), linestyle='--', color='red')

# MEDIAN STEPS BY WORK_OR_WEEKEND
# Group by the dummy variable work_or_weekend and aggregate the average of the steps
# Create a bar graph with its title and labels
# Indicate the minum as in standard vision the difference is too small to be preceived
steps_work = round(fitbit.groupby('Work_or_Weekend')['Steps'].mean(), 2)
steps_work.plot(kind = 'barh', ax=ax[1], color=['royalblue', 'forestgreen'])
ax[1].set_title('Average Steps Work vs Weekend', fontsize=15, fontweight='bold')
ax[1].set_xlabel('Avg Steps')
ax[1].set_ylabel('Day Type')
ax[1].set_xlim(left=steps_work.min()-steps_work.min()/20)

fig.savefig('../figures/figure_challenge_2.png')
```

#### Write a loop to plot 3 scatter plots of the following features:

* Minutes Lightly Active vs Steps    
* Minutes Very Active vs Steps    
* Minutes Sedentary vs Steps  

```python
# Encapsule columns of interest
# N subplots where N = length of columns of interest
# Loop thorugh the length of the subplots to edit all the axes
# Include in each axe the graph, title and labels

plots = ['Minutes Lightly Active', 'Minutes Very Active', 'Minutes Sedentary']
fig, ax = plt.subplots(1, len(plots), figsize=(25,6))

for i in range(len(plots)):
    ax[i].plot(fitbit.groupby(plots[i])['Steps'].mean())
    ax[i].set_title(plots[i], fontsize=20, fontweight='bold')
    ax[i].set_ylabel('Steps')
    ax[i].set_xlabel(plots[i])
```
